Посчитайте сумму населения по типам населенных пунктов в абсолютных и относительных значениях.

```python
df.groupby(['type'])['population'].sum().sort_values(ascending=False).reset_index() #Сумма в абсолютных значениях
```

```python
#Сумма в относительных значениях (%)

df.insert(10, 'relative_value', None)
all_population = df['population']. sum ()
df['relative_value'] = df['population'] * 100 / all_population
df.groupby('type')['relative_value'].sum().sort_values(ascending=False).reset_index().round(2)
```

Вывод: 67,88% населения живут в городах. В абсолютном значении: 102 902 660 людей

```python
#Наиболее часто встречаются у нас "Ям" и "Яр"

first_region = pd.DataFrame({'Яр': df.loc[df['only_name'] == 'Яр']['region_name'].unique()}) 
#df.loc[df['only_name'] == 'Яр' - для каждого значения столбца only_name найдем те строки, где содержится "Яр"
# далее для этих значений выводим соответствующие им уникальные данные региона (региональную принадлежность населенных пунктов)

second_region = pd.DataFrame({'Ям': df.loc[df['only_name'] == 'Ям']['region_name'].unique()})
#то же самое, но для 'Яра'

two_regions = second_region.join(first_region)
#объединяем первый датафрейм и второй
two_regions
```

```python
# Получите список населенных пунктов с длиной в три символа, найдите наиболее популярные названия из найденного списка.

df.loc[df['only_name'].str.len() == 3][['only_name']].value_counts().head(10)

#аналогично
```

```python
#Определите региональную принадлежность для наиблее часто встречающихся названий.

#Наиболее часто встречаются у нас "Бор", 'Дор' и "Мыс"

#Составим датафрейм для Бора
three_region = pd.DataFrame({'Бор': df.loc[df['only_name'] == 'Бор']['region_name'].unique()}) 

#Составим датафрейм для Дора
three_region_2 = pd.DataFrame({'Дор': df.loc[df['only_name'] == 'Дор']['region_name'].unique()})

# Объединяем первый и второй датафреймы
three_region_2 = three_region_2.join(three_region)

# Составим датафрейм бля Мыса
three_region_3 = pd.DataFrame({'Мыс': df.loc[df['only_name'] == 'Мыс']['region_name'].unique()})

#Объединяем третий датафрейм с остальными
three_region_3 = three_region_3.join(three_region_2)
three_region_3
```
